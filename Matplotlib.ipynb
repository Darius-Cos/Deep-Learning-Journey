**1.plt.plot()- În Deep Learning, o folosesti 99% din timp pentru Loss Curves (sa vezi dacamodelul invata).**

Parametri cheie de stiut:

label='Nume': Eticheta care apare în legenda.

linestyle='--': Face linia punctata (util pentru a diferenția Validarea de Antrenament).

color='red': Specifici culoarea manual.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
epoci = [1, 2, 3, 4, 5]
loss_train = [0.8, 0.6, 0.4, 0.3, 0.2]    
loss_val =   [0.9, 0.7, 0.6, 0.65, 0.8]

plt.figure(figsize=(7,5))

plt.plot(epoci,loss_train,label="Training loss",color='b',linewidth='2')
plt.plot(epoci,loss_val,label="Validation loss",color='r',linestyle='--')

plt.title('Dectare overfitting', fontsize=18)
plt.xlabel('Epoci')
plt.ylabel('Eroare',fontsize=12)
plt.grid(True)

plt.legend()
plt.show()
```

**2. plt.scatter() – Pentru Relatii și Distributii**

Parametri cheie de stiut:

c=y: inseamna "Coloreaza punctele automat în functie de valoarea lor (clasa lor)". Fara asta, toate punctele ar avea aceeasi culoare si nu ai distinge clasele.

cmap=plt.cm.RdYlBu: Alege o paleta de culori (Roșu-Galben-Albastru).

alpha=0.5: Transparenta. Daca punctele se suprapun, transparenta te ajuta sa vezi densitatea.

```python
x=np.random.rand(100,2)
y=(x[:,0]+x[:,1]>1).astype(int)

plt.figure(figsize=(7,5))

plt.scatter(x[:,0],x[:,1],c=y,s=100,cmap=plt.cm.coolwarm,edgecolor='black')
plt.show()
```

**plt.subplots() – Multi-tasking (Grid)**

Când vrei sa compari mai multe imagini deodată

```python
# 1 rand, 3 coloane
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
# 'axes' este acum o lista: [grafic1, grafic2, grafic3]

# 2. Umplem primul grafic
axes[0].plot([1, 2, 3], [1, 4, 9])
axes[0].set_title("Grafic Liniar")

# 3. Umplem al doilea grafic (Scatter)
axes[1].scatter([1, 2, 3], [1, 4, 9], color='red')
axes[1].set_title("Grafic Scatter")

# 4. Umplem al treilea grafic (Imagine)
axes[2].imshow(np.random.rand(10, 10), cmap='viridis')
axes[2].set_title("Grafic Imagine")
axes[2].axis('off')

plt.show()
```
