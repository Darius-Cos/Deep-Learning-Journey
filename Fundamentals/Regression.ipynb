## Split data into training/test set

One of the other most common and important steps in a machine learning project is creating a training and test set (and when required, a validation set).

Each set serves a specific purpose:
* **Training set** - the model learns from this data, which is typically 70-80% of the total data available (like the course materials you study during the semester).
* **Validation set** - the model gets tuned on this data, which is typically 10-15% of the total data available (like the practice exam you take before the final exam).
* **Test set** - the model gets evaluated on this data to test what it has learned, it's typically 10-15% of the total data available 


```python
len(X)
X_train=X[:40] #fist 40 samples (80% of the data) 
y_train=y[:40]

X_test=X[40:] # the last 10 samples
y_test=y[40:]

X_train,y_train,X_test,y_test
```

```python
#create de model 
model=tf.keras.Sequential([
    tf.keras.layers.Dense(50,activation=None), #hidden layer
    tf.keras.layers.Dense(1) #output layer
])
#compile the model
model.compile(loss="mae",
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
                metrics=["mae"])

model.fit(tf.expand_dims(X,axis=-1),y,epochs=100)
```

```python
model.predict(tf.expand_dims([17.0], axis=-1))
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X_train,y_train,color="b",label="Training data")
plt.scatter(X_test,y_test,c="g",label="Testing data")
plt.legend()
```

```python
model.summary()
```

```python
y_preds = model.predict(X_test)
y_preds
```

```python
#Plotting function
def plot_prediction(
        train_data=X_train,
        train_labels=y_train,
        test_data=X_test,
        test_labels=y_test,
        predictions=y_preds
):
    plt.figure(figsize=(10,7))
    plt.scatter(train_data,train_labels,color='blue',label="Training_Data")
    plt.scatter(test_data,test_labels,color='green',label="Testing_Data")
    plt.scatter(test_data,predictions,c="red",label="Prediction_Data")
    plt.legend()

    
```

```python
plot_prediction(X_train,y_train,X_test,y_test,y_preds)
```

Two of the main metrics used for regression problems are:
* **Mean absolute error (MAE)** - the mean difference between each of the predictions.
* **Mean squared error (MSE)** - the squared mean difference between of the predictions (use if larger errors are more detrimental than smaller errors).

```python
model.evaluate(X_test, y_test)
```

```python
def mae(y_test,y_preds):
    return tf.metrics.MAE(y_test,y_preds)

def mse(y_test,y_preds):
    return tf.metrics.MSE(y_test,y_preds)
```

```python
tf.random.set_seed(42)

model_1=tf.keras.Sequential([
    tf.keras.layers.Input(shape=[1]),
    tf.keras.layers.Dense(1,name='Output_layer')
]) 

model_1.compile(loss=tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.SGD(),
                metrics=['mae'],
                )
model_1.fit(X_train,y_train,epochs=100)
```

```python
# Verifică concret:
weights, biases = model_1.layers[0].get_weights()
print(f"Weight (w): {weights[0][0]:.4f}")  # Primul parametru
print(f"Bias (b): {biases[0]:.4f}")        # Al doilea parametru
```

```python
y_preds_1 = model_1.predict( X_test)
plot_prediction(predictions=y_preds_1.squeeze())
```

```python
mae_1 = mae(y_test, y_preds_1.squeeze()).numpy()
mse_1 = mse(y_test, y_preds_1.squeeze()).numpy()
print(f"Model 1 - MAE: {mae_1:.2f}, MSE: {mse_1:.2f}")
```

```python
model_2=tf.keras.Sequential([
    tf.keras.layers.Input(shape=[1]),
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(1)
])

model_2.compile(loss='mae',
                  optimizer=tf.keras.optimizers.SGD(),
                  metrics=['mae'])
model_2.fit(X_train,y_train,epochs=100)
```

```python
# Make and plot predictions for model_2
y_preds_2 = model_2.predict(X_test)
plot_prediction(predictions=y_preds_2.squeeze())
```

```python
# Calculate model_2 metrics
mae_2 = mae(y_test, y_preds_2.squeeze()).numpy()
mse_2 = mse(y_test, y_preds_2.squeeze()).numpy()
print(f"Model 2 - MAE: {mae_2:.4f}, MSE: {mse_2:.4f}")
model_2.summary()
```
